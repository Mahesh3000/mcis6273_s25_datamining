**DATA TRANSFORMATION**

There are three things
we are going to do to the data:

1. We are going to transform the `"Cacao Percent"` column to
   a floating between 0 and 1, so that the original
   string `"76%"` is tranformed to the number `0.76`.  Later
   this will facilitate machine learning algorithms.
2. We are going to split the `"Ingredients"` column 
   into multiple columns.  You will notice that
   the column currently might look something like
   `"3- B,S,C"`.  This means there are `3`
   ingredients, `"B"`, `"S"` and `"C"`.  Above the 
   table on the site is a description of what these are.
   You will need the following tools: `DataFrame.apply()`,
   `Series.str.split()` and `Series.str.strip()`.  Once
   your munging is done, make sure the column type is
   set to `int` using `pd.astype()` and that
   the resulting columns are merged into the main 
   DataFrame. _This step should
   add 8 new columns to your DataFrame: `"B", "C","L","S","S*","Sa","V"`
   and `"ingredient_count"`_.
3. We are going to split out the `"Most Memorable Characteristics"`
   into **all** of the characterstics listed which 
   occur **20 or more times**.  To do this, you 
   may need to study `Series.str.get_dummies()`.  You might
   also find the `Series.agg("sum")`, `Series.sort_values()`, `Series.where()` or `DataFrame.query()`
   of great use.  There are other ways to 
   accomplish the goal, but these are 
   accessible and easily implemented in one or two lines.  You will end up 
   with 71 new columns.


````python
import pandas as pd

df = pd.read_csv("./data/2024_flavors_of_cacoa.tsv", delimiter='\t')

#1. Convert 'Cocoa Percent' from a string to a float
df['Cocoa Percent'] = df['Cocoa Percent'].str.replace('%', '').astype(float) / 100

# Check total columns before transformation
print(f"Total columns before transformation: {df.shape[1]}")

# 2.Extracting the ingredient count
df['ingredient_count'] = df['Ingredients'].str.extract(r"^(\d+)").fillna(0).astype(int)

# Extracting the actual ingredients (everything after '-')
df['ingredients_list'] = df['Ingredients'].str.split('-').str[1]

# Defining possible ingredient types
ingredient_types = ["B", "C", "L", "S", "S*", "Sa", "V"]


for ingredient in ingredient_types:
    df[ingredient] = 0

def mark_ingredients(row):
    if pd.notna(row): # Checking if the row has a valid (non-null) value before proceeding
        present_ingredients = {i.strip() for i in row.split(',')}  # Remove spaces
        return [1 if ing in present_ingredients else 0 for ing in ingredient_types] # creating list where 1 means there 0 means not there
    return [0] * len(ingredient_types)

df[ingredient_types] = df['ingredients_list'].apply(mark_ingredients).apply(pd.Series) #Converting the returned list into separate DataFrame columns


# 3.Converting the column into one-hot encoded format
char_dummies = df['Most Memorable Characteristics'].str.get_dummies(sep=', ')

# Keeping characteristics that occur at least 20 times
frequent_chars = char_dummies.sum().sort_values(ascending=False)
selected_chars = frequent_chars[frequent_chars >= 20].index


# Adding these columns to the main DataFrame
df = pd.concat([df, char_dummies[selected_chars]], axis=1)

# Exporting the cleaned data to a CSV file
df.to_csv('data/cleaned_data_2025_flavors_of_cacao.csv', index=False)

# Extracting relevant columns for reduced data and exporting to data folder
reduced_df = df[['Review Date', 'Country of Bean Origin', 'Cocoa Percent', 'Rating']]
reduced_df.to_csv('data/data_reduced_2025_flavors_of_cacao.csv', index=False)
reduced_df.to_json('data/data_reduced_2025_flavors_of_cacao.json', orient='records', lines=True)


# Filtering data based on the criteria
filtered_df = df[
    (df['Rating'] >= 3.25) & 
    (df['Cocoa Percent'] >= 0.65) & 
    (df['Cocoa Percent'] <= 0.75) &
    (df['Review Date'].between(2018, 2021)) &
    (df['Most Memorable Characteristics'].str.contains('fatty|earthy|roasty', case=False, na=False))
]



# Extracting relevant columns for reduced data to csv and json and exporting to data folder
filtered_df.to_csv('data/data_filtered_2025_flavors_of_cacao.csv', index=False)
filtered_df.to_json('data/data_filtered_2025_flavors_of_cacao.json', orient='records', lines=True)

# Drop unnecessary columns
df.drop(columns=['ingredients_list', 'Ingredients', 'Most Memorable Characteristics'], inplace=True)

# checking the number of columns
print(f"Total columns after transformation: {df.shape[1]}")
````
